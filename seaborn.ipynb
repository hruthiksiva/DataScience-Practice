```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

```python
crash_df=sns.load_dataset('car_crashes')
crash_df.head()
```

Univariate - 1 set

```python
sns.displot(crash_df['speeding'])
```

```python
sns.distplot(crash_df['not_distracted'], kde=False, bins=25)
```

```python
#joint plot
sns.jointplot(x='speeding', y='alcohol',data=crash_df,kind='kde')
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

```python
#pairplot
sns.pairplot(crash_df)
```

```python
tips_df=sns.load_dataset('tips')
sns.pairplot(tips_df,hue='sex',palette='Blues')
```

```python
#rugplot
sns.rugplot(tips_df['tip'])
```

```python
#styling
sns.set_style('whitegrid')  #white,whitegrid,dark,darkgrid,ticks
sns.set_context('paper',font_scale=1.4) #paper, talk, poster
#sns.despine(left=False,bottom=True,top=True)
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

```python
#categorical plots
#1.1 bar plot
sns.barplot(x='sex',y='total_bill',data=tips_df,estimator=np.median)# mean, median,std,var,cov, own functions aggreatation
```

```python
#1.2 count plot
sns.countplot(x='sex', data=tips_df)
```

```python
#1.3 box plot
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')
plt.legend(loc=0) # remove scales
```

```python
#1.4 violin plot
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex',split=True)
```

```python
#1.5 Strip plot
plt.figure(figsize=(8,5))
sns.stripplot(x='day',y='total_bill',data=tips_df,jitter=True,hue='sex',dodge=True)
```

```python
#swarm Plot
sns.violinplot(x='day',y='total_bill',data=tips_df)
sns.swarmplot(x='day',y='total_bill',data=tips_df,color='white')
```
